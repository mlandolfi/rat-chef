```python
import pandas as pd
from scipy.stats import linregress
from statistics import stdev
```

```python
data = pd.read_csv('BTC-USD.txt', sep=",", header=None)
data.columns = ["time", "low", "high", "open", "close", "volume"]
data
```

```python
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l']
toMake = []
for label in labels:
    for i in range(0, 5):
        toMake.append(label+str(i))
toMake.append('slope')
toMake.append('rValue')
toMake.append('pValue')
toMake.append('label')
od = pd.DataFrame(columns=toMake)
od
```

```python
values = data['open']
maxOut = 5000
x = range(1, 61)
for i in range(0, maxOut):
    nextRow = []
    for j in range(i, i+60):
        nextRow.append(values[j])
    bois = linregress(x, nextRow)
    nextRow.append(bois.slope)
    nextRow.append(bois.rvalue)
    nextRow.append(bois.pvalue)
    nextRow.append('no' if values[i+120] < nextRow[59] else 'yes')
    od.loc[i] = nextRow
od
```

```python
slopes = od.slope
labels = od.label
slopeError = 0.1
```

```python
posCount = 0
negCount = 0
correctCount = 0
totalCount = 1
for i in range(maxOut):
    myRow = []
    for j in range(i, i+60):
        myRow.append(values[j])
    mySlope = linregress(x, myRow).slope
    for j in range(len(slopes)):
        if (abs(slopes[j]-mySlope) <= slopeError):
            if (labels[j] == 'yes'):
                posCount += 1
            else:
                negCount += 1
#     print (posCount)
#     print (negCount)
    if (posCount >= negCount and values[i] <= values[i+120]):
        correctCount += 1
        print ('\r'+str(correctCount/totalCount), end="")
    elif (negCount > posCount and values[i] > values[i+120]):
        correctCount += 1
        print ('\r'+str(correctCount/totalCount), end="")
    posCount = 0
    negCount = 0
    totalCount += 1
    
correctCount    
```
