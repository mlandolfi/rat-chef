```python
import pandas as pd
from scipy.stats import linregress
```

```python
data = pd.read_csv('BTC-USD.txt', sep=",", header=None)
data.columns = ["time", "low", "high", "open", "close", "volume"]
data
```

```python
bois = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l']
toMake = []
for boi in bois:
    for i in range(0, 5):
        toMake.append(boi+str(i))
toMake.append('label')
od = pd.DataFrame(columns=toMake)
od
```

```python
print (len(data.index)//60)
for i in range(0, len(data.index)//60):
    print ('\r'+str(i), end='')
    row = []
    for j in range(i, i+60, 5):
        temp = data.loc[j]
        row  = row + [float(temp.low), float(temp.high), float(temp.open), float(temp.close), float(temp.volume)]
    label = 'yes' if row[len(row)-2] < data.loc[i+120]['close'] else 'no'
    row.append(label)
    od.loc[i] = row
od
```

```python
newMake = toMake
newMake.pop()
maxValues = pd.DataFrame(columns=newMake)
bigBois = []
for i in range(0, len(newMake)):
    bigBois.append(od.iloc[:][newMake[i]].max())
maxValues.loc[0] = bigBois
maxValues
```

```python
odNormal = od.copy()
odNormal
```

```python
maxSeries = pd.Series(maxValues.loc[0, 'a0':'l4'].tolist())
for i in range(0, len(odNormal)):
    row = pd.Series(odNormal.loc[i, 'a0':'l4'].tolist())
    divi = row.divide(maxSeries).tolist()
    divi.append(odNormal.loc[i]['label'])
    odNormal.loc[i] = divi
odNormal
```

```python
weights = pd.Series([0.0]*60)
weights
```

```python
for i in range(0, len(odNormal.index)//2):
    row = pd.Series(odNormal.loc[i, 'a0':'l4'].tolist())
    label = odNormal.loc[i].label
    predicted = row.dot(weights)
    if (predicted >= 0 and label == 'no'):
        weights = weights.subtract(row)
    elif (predicted <= 0 and label == 'yes'):
        weights = weights.add(row)
weights
```

```python
total = 0
correct = 0
for i in range(len(odNormal.index)//2, len(odNormal.index)):
    total += 1
    row = pd.Series(odNormal.loc[i, 'a0':'l4'].tolist())
    label = odNormal.loc[i].label
    predicted = row.dot(weights)
    if (predicted >= 0 and label == 'yes'):
        correct += 1
    elif (predicted <= 0 and label == 'no'):
        correct += 1
        
round(correct/total, 3)*100
```
